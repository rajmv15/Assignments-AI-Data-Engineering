# Problem 1


````python
import numpy as np
import matplotlib.pyplot as plt

number_rice = []
total = []


for day in range (1, 100):
    number_rice.append(2 ** (day - 1))
    total.append(sum(number_rice))
    


print("The total amount of grain of rice is: {}".format(total[-1]))
````

````python
%matplotlib inline 

plt.title( "Graph of grains of rice on 100th day " )
plt.xlabel( "Number of days " )
plt.ylabel( "NUmber of grains " )
plt.plot (total, color = 'orange', label='Total',linewidth=5)
plt.plot (number_rice, color = 'blue', label='Daily', linewidth=3)
plt.legend()
plt.show()
````

# Problem 2


````python
def compute_sorori_shinzaemon(days):
    list_numb_rice = number_rice [: days]
    list_total = total [: days]

    days = range(1, days + 1) 
    return list_numb_rice, list_total

list_numb_rice30, list_total30 = compute_sorori_shinzaemon(30)
list_numb_rice81, list_total81 = compute_sorori_shinzaemon(81)


print ("The amount of grains to be received on the 30th day is: {} grains.".format(list_numb_rice30[-1]))
print ("Total number of rice gains after 30 days is: {} grains.".format(np.sum(list_numb_rice30)))

print ("The amount of grains to be received on the 81st day is: {} grains.".format(list_numb_rice81[-1]))
print ("Total number of rice gains after 81 days is: {} grains.".format(np.sum(list_numb_rice81)))
````

````python

plt.title( "Graph of grains of rice on 30th day " )
plt.xlabel( "Number of days " )
plt.ylabel( "NUmber of grains " )
plt.plot (list_total30, color = 'orange', label='Total',linewidth=5)
plt.plot (list_numb_rice30, color = 'blue', label='Daily', linewidth=3)
plt.legend()
plt.show()


plt.title( "Graph of grains of rice on 81st day " )
plt.xlabel( "Number of days " )
plt.ylabel( "NUmber of grains " )
plt.plot (list_total81, color = 'orange', label='Total',linewidth=5)
plt.plot (list_numb_rice81, color = 'blue', label='Daily', linewidth=3)
plt.legend()
plt.show()
````

# Problem 3


  Notes:
```    ------
    Based on research, an average person needs approximately 5,000 grains of rice per day.
    Source: Nutritional guidelines suggesting 100g (1/2 cup) of uncooked rice per day,
    which converts to roughly 5,000 grains.

````python
rice_per_day = 5000 #grains
number_ppl = 250 #people
date = 100 

def rice_sustainability (total_rice_grains, number_ppl):
    total_daily_consumption = number_ppl * rice_per_day
    days_livable = total_rice_grains // total_daily_consumption
    
    return days_livable


amount_of_rice = total [: date]
days_livable = rice_sustainability (amount_of_rice[-1], number_ppl)

print ("{} people can survive {} day(s) with the {} of grains of rice".format(number_ppl, days_livable, amount_of_rice[-1]))
````
