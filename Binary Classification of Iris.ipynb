### Accuracy (`accuracy_score`)
*Definition:* Percentage of all predictions that were correct.
Formula:

$$Accuracy = \frac{Correct Predictions}{Total Predictions}$$

______________________________________

### Precision (`precision_score`)
*Definition:* Of the points predicted as positive, how many were actually positive?
Formula:

$$Precison = \frac{True Positives}{True Positives + False Positives}$$

_________________________________________
### Recall (`recall_score`)

Definition: Of all the actual positive points, how many did we correctly find?

Formula:

$$Recall=
\frac{True Positives}
{True Positives + False Negatives}$$

_____________________________________
### F1 Score (`f1_score`)

Definition: The harmonic mean of precision and recall. Balances both.

Formula:

$$F1 =
2
×
\frac{Precision × Recall}
{Precision + Recall}
$$

_____________________________________________
________________________________________________
**Accuracy** → overall correctness.

**Precision** → “How clean are my positives?”

**Recall** → “How many positives did I catch?”

**F1** → “Balanced score between precision and recall.”

# Problem 7: Visualization

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches


def decision_region(
    X,
    y,
    model,
    step=0.01,
    title="decision region",
    xlabel="xlabel",
    ylabel="ylabel",
    target_names=["versicolor", "virginica"],
):
    """
    Draw the determination area of the model that learned binary classification with two-dimensional features.
    The background color is drawn from the estimated values of the trained model.
    The points on the scatter plot are training or validation data.

    Parameters
    ----------------
    X : ndarray, shape(n_samples, 2)
        Feature value
    y : ndarray, shape(n_samples,)
        label
    model : object
        Insert the installed model of the learned model
    step : float, (default : 0.1)
        Set the interval to calculate the estimate
    title : str
        Give the text of the graph Title
    xlabel, ylabel : str
        Give the text of the axis label
    target_names= : list of str
        Give a list of legends
    """
    # setting
    scatter_color = ["red", "blue"]
    contourf_color = ["pink", "skyblue"]
    n_class = 2

    # pred
    mesh_f0, mesh_f1 = np.meshgrid(
        np.arange(np.min(X[:, 0]) - 0.5, np.max(X[:, 0]) + 0.5, step),
        np.arange(np.min(X[:, 1]) - 0.5, np.max(X[:, 1]) + 0.5, step),
    )
    mesh = np.c_[np.ravel(mesh_f0), np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)

    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(
        mesh_f0, mesh_f1, y_pred, n_class - 1, cmap=ListedColormap(contourf_color)
    )
    plt.contour(
        mesh_f0, mesh_f1, y_pred, n_class - 1, colors="y", linewidths=3, alpha=0.5
    )
    for i, target in enumerate(np.unique(y)):
        plt.scatter(
            X[y == target][:, 0],
            X[y == target][:, 1],
            s=80,
            color=scatter_color[i],
            label=target_names[i],
            marker="o",
        )
    patches = [
        mpatches.Patch(color=scatter_color[i], label=target_names[i])
        for i in range(n_class)
    ]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()

```

```python
#evaluation
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, confusion_matrix

def ModelEvaluation(y_test, Prediction):
    accuracy = accuracy_score(y_test, Prediction)
    print(f"The accuracy of the model is: {accuracy}")
    precision = precision_score(y_test, Prediction)
    print(f"The precision of the model is: {precision}")
    recall = recall_score(y_test, Prediction)
    print(f"The recall of the model is: {recall}")
    f1 = f1_score(y_test, Prediction)
    print(f"The f1 of the model is: {f1}")
    print("")
    confusion = confusion_matrix(y_test, Prediction)
    print("The confusion matrix of the model is:")
    print(confusion)
```

```python
def decision_region_3(X, y, model, step=0.01, title='decision region', xlabel='xlabel', ylabel='ylabel', target_names=['versicolor', 'virginica']):
    # setting
    if len(target_names) == 2:
        scatter_color = ['red', 'blue']
        contourf_color = ['pink', 'skyblue']
        n_class = 2
    elif len(target_names) == 3:
        scatter_color = ['green', 'red', 'blue']
        contourf_color = ['lightgreen', 'pink', 'skyblue']
        n_class = 3
    # pred
    mesh_f0, mesh_f1  = np.meshgrid(np.arange(np.min(X[:,0])-0.5, np.max(X[:,0])+0.5, step), np.arange(np.min(X[:,1])-0.5, np.max(X[:,1])+0.5, step))
    mesh = np.c_[np.ravel(mesh_f0),np.ravel(mesh_f1)]
    y_pred = model.predict(mesh).reshape(mesh_f0.shape)
    # plot
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.contourf(mesh_f0, mesh_f1, y_pred, n_class-1, cmap=ListedColormap(contourf_color))
    plt.contour(mesh_f0, mesh_f1, y_pred, n_class-1, colors='y', linewidths=3, alpha=0.5)
    for i, target in enumerate(set(y)):
        plt.scatter(X[y==target][:, 0], X[y==target][:, 1], s=80, color=scatter_color[i], label=target_names[i], marker='o')
    patches = [mpatches.Patch(color=scatter_color[i], label=target_names[i]) for i in range(n_class)]
    plt.legend(handles=patches)
    plt.legend()
    plt.show()
```

```python
def ModelEvaluation3(y_test, Prediction):
    accuracy = accuracy_score(y_test, Prediction)
    print(f"The accuracy of the model is: {accuracy}")
    precision = precision_score(y_test, Prediction, average='macro')
    print(f"The precision of the model is: {precision}")
    recall = recall_score(y_test, Prediction, average='macro')
    print(f"The recall of the model is: {recall}")
    f1 = f1_score(y_test, Prediction, average='macro')
    print(f"The f1 of the model is: {f1}")
    print("")
    confusion = confusion_matrix(y_test, Prediction)
    print("The confusion matrix of the model is:")
    print(confusion)
```
