# Problem 6: Creating a diagram

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

sr = df["Species"].value_counts()
labels = 'setosa', 'virgiclor','virginica'
plt.title("A pie chart of the number of samples per label")
sr.plot.pie(autopct="%.1f%%", labels=labels);
```

```python
fig, axes = plt.subplots(1, 2)
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
df['species'] = pd.Series(iris.target)
# --- End of Setup ---

sns.boxplot(x="species", y="sepal_width",data=df, ax=axes[0])
sns.boxplot(x="species", y="sepal_width",data=df, ax=axes[1])

sns.boxplot(x="species", y="petal_width", data=df, ax=axes[0])
axes[0].set_title('Petal Length by Species')

sns.boxplot(x="species", y="petal_width", data=df, ax=axes[1])
axes[1].set_title('Sepal Length by Species')

fig.set_size_inches(12, 5)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2)

sns.violinplot(x="species", y="sepal_length",data=df,ax=axes[0])
sns.violinplot(x="species", y="petal_length",data=df,ax=axes[1])

axes[0].set_xticklabels(['setosa','virgiclor','virginica'])
axes[0].set_title('sepal_length violinplot')

axes[1].set_xticklabels(['setosa','virgiclor','virginica'])
axes[1].set_title('petal_length violinplot')

fig.set_size_inches(14, 5)
plt.show()
```

# Problem 7: Checking the relationship between feature values

```python

fig, ax = plt.subplots(2, 3, figsize=(15,10))

fig.suptitle('- Create scatter plots of all combinations (6) such that 2 out of 4 features are selected by color.')

sns.scatterplot(ax=ax[0,0], data=df, x='sepal_length', y='sepal_width', hue='species', palette="deep")
sns.scatterplot(ax=ax[0,1], data=df, x='sepal_length', y='petal_length', hue='species', palette="deep")
sns.scatterplot(ax=ax[0,2], data=df, x='sepal_length', y='petal_width', hue='species', palette="deep")
sns.scatterplot(ax=ax[1,0], data=df, x='sepal_width', y='petal_length', hue='species', palette="deep")
sns.scatterplot(ax=ax[1,1], data=df, x='sepal_width', y='petal_width', hue='species', palette="deep")
sns.scatterplot(ax=ax[1,2], data=df, x='petal_length', y='petal_width', hue='species', palette="deep")

sns.set()
     
```

```python
rel=sns.pairplot(df, hue='species')

rel.fig.subplots_adjust(top=.9)
rel.fig.suptitle('- Create a scatterplot matrix that displays all feature combinations at once.')

sns.set()
```

# Problem 8: Discussion of the results

- **On Species Variability**: The Iris virginica species exhibits the greatest variability in sepal length, petal length, and petal width, as indicated by the widest range in these measurements compared to the other species.
- **On Interspecies Trends**: A clear ranking exists for petal dimensions across the species. On average, Iris setosa has the smallest petal length and width, followed by Iris versicolor, with Iris virginica having the largest.
- **On Feature Overlap**: The distributions for sepal width overlap significantly among the three species. For instance, a sepal width between 2.0 cm and 3.5 cm is observed in samples of all three species, suggesting this feature alone is a less reliable differentiator.
- **On Feature Correlation**: A strong, positive linear correlation exists between petal length and petal width, with a correlation coefficient of 0.96. This value, being extremely close to 1, indicates that as one measurement increases, the other tends to increase proportionally.
