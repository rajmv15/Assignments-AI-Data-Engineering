# Solar System

```python
chestnut_vol = 0.00001663 #cubic meter ---> Source: Google
solarsystem_vol = 1.13*(10**38) #cubic meter ---> Source: Google
time_stamp = 0

new_vol = chestnut_vol
while new_vol <= solarsystem_vol:
    time_stamp += 1
    new_vol = chestnut_vol * (2 ** time_stamp)
    
time_stamp = time_stamp * 5

print("The time needed to cover the solar system was {}min" .format(time_stamp))
print("Or simply {:.2f}h" .format(time_stamp/60))
```

# Tokyo Dome

```python
tokyodome_vol = 1240000 #cubic meter ----> Source: https://en.wikipedia.org/wiki/Tokyo_Dome
chestnut_vol = 0.00001663
time_stamp = 0
time = 0
minutes = []
chestnut_fill = []

new_vol = chestnut_vol * (2 ** time_stamp)
while new_vol <= tokyodome_vol:
    time_stamp += 1
    new_vol = chestnut_vol * (2 ** time_stamp)
    minutes.append (time_stamp)
    chestnut_fill.append (new_vol)

import matplotlib.pyplot as plt
    
plt.plot(minutes, chestnut_fill, linewidth=2, markersize=12, markeredgewidth=3, markeredgecolor='orange', color="black", label="Time taken to fill up Tokyo Dome when you sprin")
plt.xlabel('Time in minutes')
plt.ylabel('Size of the object')
plt.title("Relating to Tokyo Dome Situation")
plt.legend(loc='upper center')
print("it takes {} Minutes".format(time_stamp)) #since it dublicates each five minutes
    
```
