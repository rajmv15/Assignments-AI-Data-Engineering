## $$ y = 12x+1 $$


# Problem 1: Linear Function

```python
import numpy as np

x_ndarray = np.arange(-50,50.1,.1)
y_ndarray = 12*x_ndarray + 1
y_ndarray, x_ndarray
```

# Problem 2: Array Combination

```python
x_ndarray = x_ndarray.reshape(-1,1)
y_ndarray = y_ndarray.reshape(-1,1)
```

```python
combined_ndarray = np.block([x_ndarray, y_ndarray])
combined_ndarray
```

# Problem 3: Find the Gradient

```python
Xn = x_ndarray[1:]
xn = x_ndarray[-1:]
Yn = y_ndarray[1:]
yn = y_ndarray[-1:]
```

```python
dx = np.array(xn - Xn)
dy = np.array(yn - Yn)

# to remove the zeros, because 0 divided by 0 creates a mathematical error
dx = dx[dx != 0]
dy = dy[dy != 0]

```

```python
gradient = dx/dy
gradient = gradient.reshape(-1,1)
print(f"The shape of the gradient matrix is {gradient.shape}")
print(gradient)
```

# Problem 5: Python functionalization

```python

def compute_gradient(function, x_range = (-50, 50.1, 0.1)):
    x_array = np.arange(*x_range)
    y_array = function(x_array)
    
    xy_array = np.concatenate((x_array[:,np.newaxis], y_array[:,np.newaxis]), axis=1)
    gradient = (xy_array[1:, 1] - xy_array[:-1, 1])/(xy_array[1:,0] - xy_array[:-1, 0])
    return xy_array, gradient

def function1(x_array):
    y_array = x_array**2
    return y_array

xy_array1, gradient1 = compute_gradient(function1)
print(compute_gradient(function1))
```

```python

plt.title("y=x**2")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(xy_array1[:,0], xy_array1[:,1],color='orange',linewidth=5)
plt.show()

plt.title("gradient of y=x**2")
plt.xlabel("X")
plt.ylabel("gradient")
plt.plot(xy_array1[:-1,0], gradient1,color='violet',linewidth=5)
plt.show()
```
