# Problem 1: Creating random numbers

```python
import numpy as np

# Seed for reproducibility
np.random.seed(0)

# The mean vector and covariance matrix
mean_vector1 = [-3, 0]
covariance_matrix1 = [[1.0, 0.8], 
                     [0.8, 1.0]]
sample_size = 500

# Generate the 500 random numbers
random_numbers = np.random.multivariate_normal(mean_vector1, covariance_matrix1, sample_size)
```

```python

print(f"Shape of the generated array: {random_numbers.shape}")
print("First 10 generated numbers:")
print(random_numbers[:10]) #printing only the first 10
```

# Problem 2: Visualization by scatter plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

x_values = random_numbers[0:500, 0]
y_values = random_numbers[0:500, 1]

plt.xlabel("x coordinate")
plt.ylabel("y coordinate")
plt.title("Scatter")
plt.scatter(x_values, y_values)
plt.show()
```

# Problem 3: Visualization by histogram

```python
counts, bins = np.histogram(x_values)
plt.hist(bins[:-1], bins, weights=counts)

plt.xlabel("x coordinate")
plt.ylabel("y coordinate")
plt.title("Scatter")
plt.hist(bins[:-1], bins, weights=counts)
plt.xlim(right=5)  # adjust the right leaving left unchanged
plt.xlim(left=-10)  # adjust the left leaving right unchanged
plt.show()
```

```python
counts, bins = np.histogram(y_values)
plt.hist(bins[:-1], bins, weights=counts)

plt.xlabel("x coordinate")
plt.ylabel("y coordinate")
plt.title("Scatter")
plt.hist(bins[:-1], bins, weights=counts)
plt.xlim(right=5)  # adjust the right leaving left unchanged
plt.xlim(left=-10)  # adjust the left leaving right unchanged
plt.show()
```

# Problem 4: Addition of data

```python
# Seed for reproducibility
np.random.seed(0)

# The mean vector and covariance matrix
mean_vector1 = [-3, 0]
mean_vector2 = [0, -3]
covariance_matrix = [[1.0, 0.8], 
                     [0.8, 1.0]]
sample_size = 500

# Generate the 500 random numbers
random_numbers1 = np.random.multivariate_normal(mean_vector1, covariance_matrix, sample_size)
random_numbers2 = np.random.multivariate_normal(mean_vector2, covariance_matrix, sample_size)
```

```python
x_values0 = random_numbers1[0:500, 0]
y_values0 = random_numbers1[0:500, 1]
x_values1 = random_numbers2[0:500, 0]
y_values1 = random_numbers2[0:500, 1]

plt.xlabel("x coordinate")
plt.ylabel("y coordinate")
plt.title("Scatter")
plt.scatter(x_values0, y_values0, alpha = .5)
plt.scatter(x_values1, y_values1, alpha = .5)
plt.show()
```

# Problem 5: Data Combination

```python
combined_randoms = np.concatenate([random_numbers1, random_numbers2])

print(f"Shape of the generated array: {combined_randoms.shape}")
```
